```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = load_digits().data
target = load_digits().target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size= 0.2)
```

```python
ks = range(1,15)
clf = GridSearchCV(KNeighborsClassifier(),{
    'n_neighbors': list(ks),
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size': [3,10,30,60],
    'weights': ['uniform', 'distance']
}, cv=5, return_train_score=False)
```

```python
clf.fit(X_train, y_train)
```

```python
pd.DataFrame(clf.cv_results_)
```

```python
results = pd.DataFrame(clf.cv_results_)
results = results[['param_n_neighbors', 'param_algorithm', 'param_leaf_size', 'param_weights', 'mean_test_score']]
```

```python
results.sort_values('mean_test_score', ascending=False, inplace=True)
results.head(30)
#The accuracy for a lot of the parameters is the same. Top accuracy 0.987483
```
