```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from joblib import dump, load

df = pd.read_csv('canada_per_capita_income.csv')
df.rename(columns = {'per capita income (US$)': 'per_capita_income'}, inplace = True)
df.head(5)
```

```python
fig = plt.figure(figsize =(12,5))
plt.plot(df.year, df.per_capita_income)
plt.grid(True)
plt.ylabel('Per capita income')
plt.xlabel('Year')
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[['year']],df.per_capita_income, test_size = 0.1)
model = linear_model.LinearRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
dump(model, 'simple_linear_regression.joblib')
```
